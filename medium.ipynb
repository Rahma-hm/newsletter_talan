```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Part-1 : Browser Automation with selenium

```python
#open medium
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(r"C:\Users\Rahma\chromedriver\chromedriver.exe",chrome_options=chrome_options)
driver.get('https://medium.com/tag/programming')
#keywords
search = driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div/div/div[2]/div/div/div/div[1]/div[2]/div[1]/div/input')
search.click()
driver.implicitly_wait(10)
search.send_keys('Metaverse')
search.send_keys(Keys.ENTER)
driver.implicitly_wait(15)
#topic
topic = driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div/div/div[2]/div/div/div/div[1]/div[2]/div[2]/div[2]/div[1]/a/div')
topic.click()
driver.implicitly_wait(15)
#best articles
best_article = driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div/div/main/div/div[1]/div/div[2]/div/div[4]/div[1]/ul/li[3]/p/a')
best_article.click()
driver.implicitly_wait(15)
#last week
time = driver.find_element_by_xpath('//*[@id="root"]/div/div[3]/div/div/main/div/div[1]/div/div[2]/div/div[4]/div[3]/button')
time.click()
last_week = driver.find_element_by_link_text('This week')
last_week.click()
#current
current_url = driver.current_url



#****************************************************************************#


#get request 
response= requests.get(current_url)
#soup object 
soup = BeautifulSoup(response.content,'html.parser')
#result items 
list_items = soup.find_all('article')
# list article
#article title
article_title = [result.find('h2').get_text() for result in list_items]
#description
description = [result.find('div',{'class' : 'h k aj ak by'}).get_text() for result in list_items]
#link
link = [result.find('a')['href'] for result in list_items]
#image
image = [result.find('img')['src'] for result in list_items]
#author
author = [result.find('a',{'class':'au av aw ax ay az ba bb bc bd be bf bg bh bi'}).get_text() for result in list_items]
#date
date = [result.find('p',{'class':'bo b bp bq br'}).get_text() for result in list_items]
#key words

key_words = [result.find(class_ ='lo ct ht jh ji jj bo b lp bq fu jk').get_text() for result in list_items ]
#read time
read_time = [result.find(class_ ='bo b lp bq br').get_text() for result in list_items ]
#create articles df
articles = pd.DataFrame({'Title': article_title ,'Description':description, 'Link': link, 'Image': image,'Author':author,'Date':date,'Key Words':key_words,'Read time':read_time})
#output in excel
articles.to_excel('articles_medium.xlsx', index =False)
```
